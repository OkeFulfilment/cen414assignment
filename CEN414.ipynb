```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
import statistics 
import pandas

#sns.set()
```

```python
af_countries = [    'Algeria',
                    'Benin',
                    'Botswana',
                    'Burkina Faso',
                    'Burundi',
                    'Cameroon',
                    'Cape Verde', 
                    'Chad',
                    'Comoros',
                    'Congo',
                    'Dijibouti',
                    'Egypt',
                    'Eritrea',
                    'Ethiopia',
                    'Ghana',
                    'Guinea-Bissau',
                    'Malawi',
                    'Mali',
                    'Mauritania',
                    'Mauritus',
                    'Morocco',
                    'Niger',
                    'Senegal',
                    'Seychelles',
                    'South Africa',
                    'Sudan',
                    'Tanzania',
                    'Togo',
                    'Tunisia',
                    'Uganda',
                    'Zambia',
                    'Zimbabwe'
        ]

df   = pd.read_csv('book2.csv')
df = df[ (df['Country or Area'].isin(af_countries) )]
df = df.sort_values('Value')
df = df.set_index('Country or Area')
df
                
```

```python
for i in range(2):
    l = 1999+i*3
    h = l + 3
    fig, axes = plt.subplots(1,3, figsize=(20,10))
    
    for idx, year in enumerate(range(l,h)):
        df[df['Year'] == year]['Value'].plot.barh(ax=axes[idx])
        axes[idx].set_title(year)
```

```python
plt.figure(figsize=(10,10))

def interquartile_range(sample):
    q3, q1 = np.percentile(sample, [75 ,25])
    return q3 - q1

def quartile_range(sample):
    q3, q1 = np.percentile(sample, [75 ,25])
    return (q3 - q1)/2

def _range(sample):
    return sample.max() - sample.min()

def create_plot(function, label):
    dispersion_list = []
    for year in range(1999,2005):
            dispersion_list.append(function(df[df["Year"] == year].Value))
              
    plt.plot(range(1999,2005), dispersion_list, label=label)
    plt.title('Measures of Dispersion of African Countries against Year.')
    plt.ylabel('Value')
    plt.xlabel('Year')
    
create_plot(statistics.mean, 'Mean')
#create_plot(statistics.variance, 'Variance')
create_plot(statistics.stdev, 'Standard Deviation')
create_plot(interquartile_range, 'Interquartile Range')
create_plot(quartile_range, 'Quartile Range')
#create_plot(_range, 'Range')

plt.legend(loc='best')
```

```python
create_plot(statistics.variance, 'Variance')
plt.legend(loc='best')
```

```python
create_plot(_range, 'Range')
plt.legend(loc='best')
```
